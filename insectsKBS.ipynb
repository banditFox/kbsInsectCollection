## Data Dictionary
* Order: taxonomic order of the organism <br />
* Family: taxonomic family of the organism <br />
* Genus: taxonomic genus of the organism <br />
* Specific Epithet: taxonomic specific epithet of the organism <br />
* Author: party responsible for the formal description of the organism <br />
* CommonName: a common name of the organism <br />
* Comments: comments for this record <br />
* IdentifyingCharacteristics: identifying characteristics <br />
* CommonGroup: a common group to which this organism belongs <br />
* FunctionalGroup: a functional group to which this organism belongs <br />
* Hosts: hosts <br />
* LifeHistory: Life history <br />
* Habitat: Habitat in which this organism occurs <br />
* GeographicDistribution: Geographic distribution of this organism <br />
* Origin: Native to North America or exotic <br />
* Species: a code for identifying this species unique within the KBS namespace <br />
* Specimen: a code for uniquely identifying this specimen <br />
* StorageLocation: Where the specimen is stored; 'photo only' if specimen was photographed but not collected <br />
* Stage_Sex: Life history stage and/or sex of the specimen <br />
* Treatment: LTER treatment in which the specimen was encountered. <br /> 
* Replicate: LTER replicate plot (nested within treatment) in which the specimen was encountered. <br />
* Collector: Person who collected the specimen <br />
* img_src (Image Source): URL for a digital image of the specimen <br />
